```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read data from file '2018-2020Accidents.csv'
data = pd.read_csv("2018-2020Accidents.csv")
#data
# remove all rows with latitude or longitude of NaN
data_with_coords = data.dropna(subset=['Latitude', 'Longitude'])
#data_with_coords
```

```python
# Filter data to only include coordinates within the bounds of Monmouth County
print(data_with_coords["Latitude"].min(), data_with_coords["Longitude"].min(), data_with_coords["Latitude"].max(), data_with_coords["Longitude"].max())
correct_coords = data_with_coords[(data_with_coords["Latitude"] > 40.0886) & (data_with_coords["Latitude"] < 40.4688) & (data_with_coords["Longitude"] > 73.9685) & (data_with_coords["Longitude"] < 74.6082)]
correct_coords

# convert all longitudinal data to negative
correct_coords["Longitude"] = correct_coords["Longitude"] * -1
```

```python
print(correct_coords["Total Injured"].value_counts())
print(correct_coords["Road Grade"].value_counts())
print(correct_coords["Road Surface Type"].value_counts())
print(correct_coords["Surface Condition"].value_counts())
print(correct_coords["Light Condition"].value_counts())
```

```python
#plt.scatter(correct_coords["Total Injured"], correct_coords["Light Condition"], s=0.1)
fig = plt.figure()
plt.scatter(correct_coords["Longitude"], correct_coords["Latitude"], s=0.1)
fig.savefig('scatter.png', transparent=True)
# get starting and ending latitude and longitude of scatter plot

plt.scatter(correct_coords["Longitude"], correct_coords["Latitude"], correct_coords["Total Injured"], s=0.1)

```
